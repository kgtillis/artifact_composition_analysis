```python
## 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the CSV data into a DataFrame
data = pd.read_csv('../../data/clean_data_w_NaN.csv')

# Define a list of element names and error names
column_names = data.columns[6:].tolist()
element_names = [name for name in column_names if "Error" not in name]
error_names = [name for name in element_names if "Error" in name]

# Create subplots for each pair of Element and Error columns
num_elements = len(element_names)
fig, axes = plt.subplots(nrows=num_elements, ncols=2, figsize=(10, 50))

for i, (element, error) in enumerate(zip(element_names, error_names)):
    # Create a pivot table with 'X' and 'Y' as the index, element as columns, and values as the data
    pivot_data = data.pivot_table(index=['X', 'Y'], columns=element, values=error)

    # Create a heatmap for error values
    ax = axes[i, 0]
    sns.heatmap(pivot_data, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=0.5, ax=ax)
    ax.set_title(f'{element} Error Heatmap')

    # Create a pivot table for the element values
    pivot_data = data.pivot_table(index=['X', 'Y'], columns=element, values=element)

    # Create a heatmap for element values
    ax = axes[i, 1]
    sns.heatmap(pivot_data, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=0.5, ax=ax)
    ax.set_title(f'{element} Value Heatmap')

# Adjust spacing between subplots
plt.tight_layout()

# Show the plots
plt.show()
```
